```python
# 普通线性回归
%matplotlib inline
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

x = np.arange(100).reshape(-1, 1)
y = np.array([i*3+1+np.random.randn(1)[0]*10 for i in range(100)])
model = LinearRegression()

model.fit(x, y)

print(f'回归系数:{model.coef_}\n截距:{model.intercept_}')

new_x = np.arange(100, 110).reshape(-1, 1)
predicted_y = model.predict(new_x)
fit_y = model.predict(x)
plt.scatter(x, y)
plt.plot(x, fit_y)
# plt.scatter(new_x, predicted_y, 'o', color='red')
plt.show()
```

```python
# 最小二乘法
import numpy as np
from sklearn.linear_model import LinearRegression
%matplotlib inline
import matplotlib.pyplot as plt

x = np.arange(100).reshape(-1, 1)
y = np.arange([i*3+1+np.random.randn(1)[0]*5 for i in range(100)])
```

```python
# 岭回归
from sklearn.linear_model import Ridge
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

data = load_diabetes()
x = data.data
y = data.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

alpha = 1
ridge = Ridge(alpha=alpha) # Lasso
ridge.fit(x_train, y_train)

y_pred = ridge.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
print(f'岭回归MSE：{mse:.2f}')
```
